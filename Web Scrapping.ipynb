**Common roadblocks when web scraping**

Some of them are side effects of the functionality of the webpage.
Others are intentionally implemented by the developers to restrict the bad activities on their site.

Some of the most common issues that may arise during scrapping include:
 - Incorporation of request headers
 - Dealing with log-in systems
 - Limiting the number of requests
 - Captchas
 - Data generated by JavaScript
 
**Request headers**

When we send request to a server, we are not just asking for information, we are also providing information about the request itself. Information such as encoding and language of the expected response, length and type of data provided, the requester and cookies. These information are intented to make communication on the web easier and more reliable.

**Common fields - 'User agent and cookie'**

When a software sends a request, it identifies itself, its application type, operating system, software vendor and software version by submitting a characteristic identification (user agent string) - an ID card containing some basic information. All browsers'user agents are available publicly.  

**Cookies**

- This is a special type of request header that represents small piece of data set from a website and stored on the user's computer by the browser.
- They are designed to be a reliable mechanism for websites to remember useful information such as recording the user's activities.
- They can also be used to remember the user's information previously entered such as names, addresses, passwords and credit card numbers.
- They also perform essential functions in the modern web such as the authentication cookie which is used by web servers to know whether the user is logged in or not and which account they logged in with.

**Log-in systems**
Information that require a login to access is generally not public.
This means that distribution or commercial usage might be illegal.Always check the legality of your actions.
Depending on popularity and security of the website you are trying to access, signing in might be difficult.

Flow: Sign in - Login - Username - Password - Post request is sent - the server pocesses the data and checks its validity (incase credentials are correct).
Although, only the email and password are required upon login, the form sends additional information to the server. This often includes some authenticity token.
The information to find out are; where the 'post'request is sent to
and what format of the data. The ways to do this is either by inferring the information from the HTML file or intercept the requests the browser submits.

**Limiting requests**
Feeding a website with too many request in a short period of time is not a good practice as it can hurt the website especially a small one. The solution is to tell python to wait some amount of time before making a request.

**Captchas**
Sites that value security may include captchas to confirm that the user is not a robot. They are designed to distinguish programs so they are hard to deal with. Most websites with captchas avoid being scrapped. It is advisable not to scrape them or google how to scrape them on your own.
